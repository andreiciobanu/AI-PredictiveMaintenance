```python
import os
import time
import random
import datetime
import pandas as pd
import pytest
import fastparquet
from dateutil import parser
from ipywidgets import IntProgress
import urllib.request

from pathlib import Path
data_dir = str(Path.home()) + '/data'
%rm -rf $data_dir
%mkdir $data_dir $data_dir/logs $data_dir/telemetry

telemetry = []
errors = []
date_from = parser.parse("May 15 2017 12:00AM")
date_to = parser.parse("May 17 2018 12:00AM")
errors.append({
                'timestamp': date_from,
                'level': 'INFO',
                'code': 'fixed'
            })

errors.append({
                    'timestamp': date_from,
                    'level': 'CRITICAL',
                    'code': 'wrpmg'
                })
logs_df = pd.DataFrame(errors)
sequence_count = len(logs_df[logs_df.level == 'CRITICAL'])
logs_df.index = logs_df['timestamp']
del logs_df['timestamp']
logs_df.to_parquet(data_dir + '/logs/logs.parquet', engine='fastparquet', times='int96')
print('{0} run-to-failure sequence(s) generated.'.format(sequence_count))
print('{0}'.format(data_dir + '/logs/logs.parquet'))

with urllib.request.urlopen(data_dir + '/logs/logs.parquet') as f:
    html = f.read().decode('utf-8')
```
